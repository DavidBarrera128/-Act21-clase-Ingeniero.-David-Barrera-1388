```python
# Definición de la clase
class Ingeniero:
    # El método __init__ inicializa los atributos de cada objeto
    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre       # Atributo 1
        self.telefono = telefono   # Atributo 2
        self.edad = edad           # Atributo 3

    # Función/Acción 1
    def planificar(self, proyecto):
        print(f"{self.nombre} está planificando el proyecto: {proyecto}.")

    # Función/Acción 2
    def organizar(self, recurso):
        print(f"{self.nombre} está organizando los {recurso} del equipo.")

    # Función/Acción 3
    def programar(self, lenguaje):
        print(f"{self.nombre} está programando una solución en {lenguaje}.")

# --- Creación de Objetos (Instancias) ---

# Creando el ingeniero1 con datos de ejemplo
ingeniero1 = Ingeniero("Ana García", "555-0123", 28)

# Creando el ingeniero2 con datos de ejemplo
ingeniero2 = Ingeniero("Carlos Ruiz", "555-9876", 35)

# --- Ejecución de las funciones ---

print(f"--- Ficha de {ingeniero1.nombre} ---")
ingeniero1.planificar("Sistema de Riego Automático")
ingeniero1.organizar("horarios de mantenimiento")
ingeniero1.programar("Python")

print("\n" + "-"*30 + "\n")

print(f"--- Ficha de {ingeniero2.nombre} ---")
ingeniero2.planificar("Puente Peatonal Urbano")
ingeniero2.organizar("presupuestos de materiales")
ingeniero2.programar("C++")
```
